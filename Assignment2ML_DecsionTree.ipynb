```python



def split(splitvalue,df,col): #Function to calculate entropy and information gain for given split value
    split1 = [0,0,0] #conuters to record the number of each 
    split2 = [0,0,0]
    Total1 = 0
    Total2 = 0
    ent1 = 0
    ent2 = 0
    Ent_S = 0
    Label = df.columns.get_loc('Style')

    for i in range(len(df)):
        
        
        if df.iloc[i,col] >= splitvalue:
            Total1 += 1
            if df.iloc[i,Label] == 'ale':
                split1[0] += 1 
            
            
            elif df.iloc[i,Label] == 'lager':
                split1[1] += 1
            
        
            elif df.iloc[i,Label]  == 'stout':
                split1[2] += 1
            
            
        else:
            Total2 += 1
            if df.iloc[i,Label] == 'ale':
                split2[0] += 1 
            
            
            elif df.iloc[i,Label] == 'lager':
                split2[1] += 1
            
        
            elif df.iloc[i,Label]  == 'stout':
                split2[2] += 1
            

 

   
    for n in range(len(split1)):
        if split1[n] != 0 and Total1 != 0:
            ent1 += -split1[n]/Total1 * math.log(split1[n]/Total1,2)
            
            
        if split2[n] != 0 and Total2 != 0:
            ent2 += -split2[n]/Total2 * math.log(split2[n]/Total2,2)
            
 
    
    length = len(df)
    Ale = (split1[0]+split2[0])/length
    Lager = (split1[1]+split2[1])/length
    Stout = (split1[2]+split2[2])/length
    
    Global_ent = [Ale,Lager,Stout]
    
    for n in range(len(Global_ent)):
        if Global_ent[n] != 0:
            Ent_S += -Global_ent[n]* math.log(Global_ent[n],2) 
        
        
     
    Info_gain = Ent_S -(Total1/length * ent1) -(Total2/length * ent2) #Weighted average entropies 
    
   
    return Info_gain
    
g = traindf
g = g.drop(columns =['Nitrogen'])
split(2,g,1)
```

```python
def Evaluate_Split(df):
   
    gain = 0
    value = 0
    colname = ''
    
    for c in range(len(df.columns)): #loop through each column 
        
        if df.columns[c] == 'Style' or df.columns[c] == 'Prediction':
            print("Label")
           
            
            
        else:
            for i in range(len(df)): #loop through each row 
                currentgain = split(df.iloc[i,c],df,c) 
                if currentgain >gain:             #find value with highest info gain
                    gain = currentgain
                    value = df.iloc[i,c]          #get split value
                    colname = df.columns[c]       #get column name

         
    print("Optimal Split Point : ",value , " Gain : ",gain)
    print('-----------------------')
    return value,colname




Evaluate_Split(traindf)
```

```python
class Node():
    def __init__(self):  #class to represent tree node
        self.value = None
        self.col = None
        self.left_child = None
        self.right_child = None

    def get_name(self):
        return 'Node'
    
    def set_left(self,left):
        self.left_child = left
        
    def set_right(self,right):
        self.right_child = right
        
    def set_value(self,val,col):
        self.value = val 
        self.col = col
        
    def get_left(self):
        return self.left_child
    
    def get_right(self):
        return self.right_child
        
    def get_value(self):
        return self.value
    
    def get_col(self):
        return self.col
        

    
class Leaf():  
    def __init__(self):  # class to represent tree leaf
        self.value = None
        
    def set_leaf(self,value):
        
        self.value = value
            
    def get_value(self):
        return self.value 
        

    def get_name(self):
        return 'Leaf'
    
```

```python
def createTree(df):
    
    bestsplit = 0
    colname = ''
    
    def splitnode(df,split): #function to calculate split dataframes
        
        
        high = df.loc[df[split[1]]> split[0]]   #df with value higher than split value
        high = high.drop(columns =[split[1]])   #drop col with best info gain to calculate the next best
    
        low = df.loc[df[split[1]]<= split[0]]
        low = low.drop(columns =[split[1]])
        
        return high,low
    
    def determineLabel(df):
        labels = df.Style.unique()
        if len(labels) ==1: # One unique label return value
            return labels[0]
        
        elif len(labels) ==2: # two unique labels determine which is in the majority 
            x = (df.Style == labels[0]).sum()
            y = (df.Style == labels[1]).sum()
            
            if x/len(df)>.5:
                return labels[0]
            
            elif y/len(df)>.5:
                return labels[1]
            
            else:
                return labels[0]
            
        else:   # three unique labels determine which is in the majority, esle return random  
            x = (df.Style == labels[0]).sum()
            y = (df.Style == labels[1]).sum()
            z = (df.Style == labels[2]).sum()
            
            if x/len(df)>.5:
                return labels[0]
            
            elif y/len(df)>.5:
                return labels[1]
            
            elif z/len(df)>.5:
                return labels[2]
            
            else:
                rand = random.randint(0, 2)
                return labels[rand] 
            
            
            
        
        
    
    bestsplit = Evaluate_Split(df)
    root = Node()
    root.set_value(bestsplit[0],bestsplit[1])
    
    h,l = splitnode(df,bestsplit) #split df, this is done to evaluate the next best split value taking into account 
                                  # the split that has previously occured
    
    
 
    splitr = Evaluate_Split(h)  #find second split value for right df
    h1,l1 = splitnode(h,splitr) #split dataframe 
    
    rightnode = Node()
    rightnode.set_value(splitr[0],splitr[1])
    root.set_right(rightnode)
    
    rightleaf = Leaf()
    rightleaf.set_leaf(determineLabel(h1))
    rightnode.set_right(rightleaf)
    
    leftleaf = Leaf()
    leftleaf.set_leaf(determineLabel(l1))
    rightnode.set_left(leftleaf)
    
    
    
    splitl = Evaluate_Split(l)
    h1,l1 = splitnode(l,splitl)
    
    leftnode = Node()
    leftnode.set_value(splitl[0],splitl[1])
    root.set_left(leftnode)
    
    rightleaf2 = Leaf()
    rightleaf2.set_leaf(determineLabel(h1))
    leftnode.set_right(rightleaf2)
    
    leftleaf2 = Leaf()
    leftleaf2.set_leaf(determineLabel(l1))
    leftnode.set_left(leftleaf2)
    
    
    
    print(bestsplit,splitl,splitr)
    
    
    return root
  
    
createTree(traindf)    
```

```python
def TestTree(Tree,df):
    
    prediction = ['']*len(df)
    
    for i in range(len(df)):
        split_val = 0
        
        row = df.iloc[i]
        colname = Tree.get_col()
        col_index = df.columns.get_loc(colname)
        split_val = Tree.get_value()
        
        if row[col_index] > split_val:
            node = Tree.get_right() 
            split_val = node.get_value()
            colname = node.get_col()
            col_index = df.columns.get_loc(colname)
            
            if row[col_index] > split_val:
                node = node.get_right()
                prediction[i] = node.get_value()
            
            else:
                node = node.get_left()
                prediction[i] = node.get_value()
   
        else:
            node = Tree.get_left() 
            split_val = node.get_value()
            colname = node.get_col()
            col_index = df.columns.get_loc(colname)
            
            if row[col_index] > split_val:
                node = node.get_right()
                prediction[i] = node.get_value()
            
            else:
                node = node.get_left()
                prediction[i] = node.get_value()
        
        
    counter = 0   
    for x in range(len(df)):
        
        if df['Style'].iloc[x] == prediction[x]:
            counter+= 1
            
    
    print('')
    print('')
    print('------------------------------------------')
    print('classification rate : ',(counter/len(df))*100,'%')
    
   
    

tree1 = createTree(traindf) 
TestTree(tree1,testdf)
```
